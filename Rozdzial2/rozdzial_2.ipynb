```python
import pandas as pd
from  sqlite3 import connect#bibloteka do tworzenia połączenia z bazą danych
```

```python
#Tworzenie połączenia z bazą danych(wzięta ze strony: https://www.sqlitetutorial.net/sqlite-sample-database/)
con = connect('chinook.db')
```

```python
#Weźmy przy pomocy SQL wszystkie utwory z tabeli tracks
dataframe = pd.read_sql_query('SELECT * FROM tracks', con)

dataframe.head(100)#wyświetlmy 100 pierwszych rekordów
```

```python
#Weźmy tylko te utwory, gdzie MediaTypeId=3
#Zauważmy, że wielkość liter w słowach kluczowych nie ma znaqczenia
dataframe = pd.read_sql_query('select * from tracks where MediaTypeId=2', con)

dataframe.head(100)#wyświetlmy 100 pierwszych rekordów
```

```python
#Zróbmy jakieś fancy zapytanie sql
dataframe = pd.read_sql_query('select tracks.TrackId,tracks.Name, media_types.Name from tracks join media_types where tracks.MediaTypeId=2', con)

dataframe.head(100)#wyświetlmy 100 pierwszych rekordów
```
